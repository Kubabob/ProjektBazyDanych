```python
dane = pd.read_excel('dane harcerek.xlsx', sheet_name='daneSQL')
```

```python
dane = dane.replace('-', np.nan)
```

```python
dane.head()
```

```python
tabela_adres = dane.loc[:,['adres zamieszkania', 'kod pocztowy', 'miejscowość']].drop_duplicates('adres zamieszkania').reset_index(drop=True)
```

```python
tabela_adres
```

```python
#tabela_adres
for idx in range(tabela_adres.shape[0]):
    adres_zam, kod, miejscowosc = tabela_adres.iloc[:,0], tabela_adres.iloc[:,1], tabela_adres.iloc[:,2]
    ulica, numer = adres_zam[idx].split(' ')
    print(f"insert into adres(id_adres, ulica, numer, kod, miejscowosc) values ({idx+1}, '{ulica}', '{numer}', '{kod[idx]}', '{miejscowosc[idx]}');")
```

```python
tabela_rodzice = dane.loc[:,['imię matki', 'imię ojca', 'tel matki', 'tel ojca', 'mail do rodzica']].drop_duplicates('tel matki').reset_index(drop=True)
```

```python
for idx in range(tabela_rodzice.shape[0]):
    mimie, timie, mnumer, tnumer, mail = tabela_rodzice.iloc[:,0], tabela_rodzice.iloc[:,1], tabela_rodzice.iloc[:,2], tabela_rodzice.iloc[:,3], tabela_rodzice.iloc[:,4]
    print(f"insert into rodzice(id_rodzice, imie_matki, imie_ojca, nuer_matki, numer_ojca, mail) values ({idx+1}, '{mimie[idx]}', '{timie[idx]}', '{mnumer[idx]}', '{tnumer[idx]}', '{mail[idx]}');")
```
