```python
import pandas as pd
import numpy as np
```

```python
dane = pd.read_excel('dane harcerek.xlsx', sheet_name='daneSQL')
```

```python
dane = dane.replace('-', np.nan)
```

```python
dane.head()
```

```python
tabela_adres = dane.loc[:,['adres zamieszkania', 'kod pocztowy', 'miejscowość']].drop_duplicates('adres zamieszkania').reset_index(drop=True)
```

```python
tabela_adres
```

```python
#tabela_adres
for idx in range(tabela_adres.shape[0]):
    adres_zam, kod, miejscowosc = tabela_adres.iloc[:,0], tabela_adres.iloc[:,1], tabela_adres.iloc[:,2]
    ulica, numer = adres_zam[idx].split(' ')
    print(f"insert into adres(id_adres, ulica, numer, kod, miejscowosc) values ({idx+1}, '{ulica}', '{numer}', '{kod[idx]}', '{miejscowosc[idx]}');")
```

```python
tabela_dane = dane.loc[:, ['imię', 'nazwisko', 'PESEL', 'data urodzenia', 'adres zamieszkania']]
```

```python
np.where(tabela_adres['adres zamieszkania'] == tabela_dane.loc[idx, 'adres zamieszkania'])[0][0]
```

```python
def check_pesel(pesel):
    if len(str(pesel)) != 11:
        return f'0{pesel}'
    else:
        return str(pesel)
```

```python
for idx in range(tabela_dane.shape[0]):
    print(f"insert into podstawowe_dane(id_podstawowe_dane, id_adres, imie, nazwisko, PESEL, data_urodzenia) values ({idx+1}, {np.where(tabela_adres['adres zamieszkania'] == tabela_dane.loc[idx, 'adres zamieszkania'])[0][0]+1}, '{tabela_dane.loc[idx, 'imię']}', '{tabela_dane.loc[idx, 'nazwisko']}', '{check_pesel(tabela_dane.loc[idx, 'PESEL'])}', '{tabela_dane.loc[idx, 'data urodzenia'].split(' ')[0].replace('.', '-')}');")
```
